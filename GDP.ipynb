```python
import pandas as pd
df = pd.read_csv('Economic Indicators (WDB).csv')
df.head()
```

```python
years = [f"{year} [YR{year}]" for year in range(1970, 2024)]
relevant_columns = ["Country Name", "Series Name", "Series Code"] + years
df = df[relevant_columns]

wdb_time_series_data = pd.melt(
    df,
    id_vars=["Country Name", "Series Name", "Series Code"],
    value_vars=years,
    var_name="Year",
    value_name="Value"
)

wdb_time_series_data["Year"] = wdb_time_series_data["Year"].str.extract(r'(\d{4})').astype(int)


wdb_time_series_data["Value"] = pd.to_numeric(wdb_time_series_data["Value"], errors="coerce")

wdb_time_series_data = wdb_time_series_data.pivot_table(
    index=["Country Name", "Year"],
    columns="Series Name",
    values="Value"
).reset_index()

# Display the reshaped data
wdb_time_series_data.head()
```

```python
import yfinance as yf

sp500 = yf.download("^GSPC", start="1970-01-01", end="2023-12-31")
sp500.reset_index(inplace=True)
sp500["Year"] = sp500["Date"].dt.year
sp500["Date"] = pd.to_datetime(sp500["Date"])
sp500.set_index('Date', inplace=True)
print()
print(sp500.head())
print()
print(sp500.shape)
```

```python
sp500["Monthly_Return"] = sp500["Adj Close"].pct_change()  # Calculate monthly returns
sp500["Rolling_Mean_3M"] = sp500["Adj Close"].rolling(window=3).mean()  # 3-month rolling average
sp500["Rolling_Std_3M"] = sp500["Adj Close"].rolling(window=3).std()  # 3-month rolling standard deviation
sp500["Rolling_Mean_6M"] = sp500["Adj Close"].rolling(window=6).mean()  # 6-month rolling average
sp500["Rolling_Std_6M"] = sp500["Adj Close"].rolling(window=6).std()  # 6-month rolling standard deviation
sp500["Rolling_Mean_12M"] = sp500["Adj Close"].rolling(window=12).mean()  # 12-month rolling average
sp500["Rolling_Std_12M"] = sp500["Adj Close"].rolling(window=12).std()  # 12-month rolling standard deviation

sp500.columns = ["_".join(filter(None, col)).strip() for col in sp500.columns]
sp500.rename(columns={
    "Adj Close_^GSPC": "Adj_Close",
    "Close_^GSPC": "Close",
    "High_^GSPC": "High",
    "Low_^GSPC": "Low",
    "Open_^GSPC": "Open",
    "Volume_^GSPC": "Volume"
}, inplace=True)

sp500.columns
```

```python
sp500_yearly_data = sp500.resample('YE').agg({
    'Adj_Close': 'last',  # Last value of the year
    'Close': 'last',
    'High': 'max',  # Highest value of the year
    'Low': 'min',  # Lowest value of the year
    'Open': 'first',  # First value of the year
    'Volume': 'sum',  # Total volume for the year
    'Monthly_Return': 'mean',  # Annual average of monthly returns
    'Rolling_Mean_3M': 'last',  # Last value of rolling mean for the year
    'Rolling_Std_3M': 'last',  # Last value of rolling std for the year
    'Rolling_Mean_6M': 'last',
    'Rolling_Std_6M': 'last',
    'Rolling_Mean_12M': 'last',
    'Rolling_Std_12M': 'last'
})

sp500_yearly_data.columns = [col + '(S&P 500)' if col != 'Year' else col for col in sp500_yearly_data.columns]

sp500_yearly_data['Year'] = sp500_yearly_data.index.year

sp500_yearly_data.head()
```

```python
print(sp500_yearly_data.isnull().sum())
```

```python

wdb_sp500 = pd.merge(wdb_time_series_data, sp500_yearly_data, on='Year', how='inner')
wdb_sp500['Year'] = pd.to_datetime(wdb_sp500['Year'], format='%Y')
wdb_sp500.sort_values(by='Year', inplace=True)
wdb_sp500.set_index('Year', inplace=True)
wdb_sp500.head()
```

```python
wdb_sp500.info()
```

```python
wdb_sp500.isnull().sum()
```

```python
time_series_columns = [
    "Carbon dioxide (CO2) emissions (total) excluding LULUCF (Mt CO2e)",
    "Exports of goods and services (% of GDP)",
    "Exports of goods and services (annual % growth)",
    "Imports of goods and services (annual % growth)",
    "Life expectancy at birth, total (years)",
    "Trade (% of GDP)"
]
wdb_sp500[time_series_columns] = wdb_sp500[time_series_columns].interpolate(method='time',  limit_direction='both')
wdb_sp500[time_series_columns] = wdb_sp500[time_series_columns].ffill().bfill()

mean_columns = [
    "General government final consumption expenditure (% of GDP)",
    "General government final consumption expenditure (annual % growth)",
    "General government final consumption expenditure (current US$)",
    "Gini index"
]
wdb_sp500[mean_columns] = wdb_sp500[mean_columns].fillna(wdb_sp500[mean_columns].mean())

median_columns = [
    "Lending interest rate (%)",
    "Total natural resources rents (% of GDP)",
    "Net investment in nonfinancial assets (% of GDP)",
    "Net investment in nonfinancial assets (current LCU)",
    "Real effective exchange rate index (2010 = 100)"
]
wdb_sp500[median_columns] = wdb_sp500[median_columns].fillna(wdb_sp500[median_columns].median())

print(wdb_sp500.isna().sum())
```

```python
# Plot of GDP over the years
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(wdb_sp500.index.year, wdb_sp500['GDP (current US$)'], label='GDP (current US$)', marker = 'o')
plt.xlabel('Year')
plt.ylabel('GDP (current US$)')
plt.title('GDP Over the Years')
plt.xticks(wdb_sp500.index.year, rotation = 90)
plt.legend()
plt.grid(True)
plt.show()
```

```python
from statsmodels.tsa.stattools import adfuller

def adf_test(series):
    result = adfuller(series, autolag='AIC')
    print('ADF Statistic:', result[0])
    print('p-value:', result[1])
    if result[1] <= 0.05:
        print("Series is stationary")
    else:
        print("Series is not stationary")
    return (series.name, result[1])

adf_test(wdb_sp500['GDP (current US$)'])
```

```python
wdb_sp500['GDP_log'] = np.log(wdb_sp500['GDP (current US$)'])
wdb_sp500['GDP_log_diff'] = wdb_sp500['GDP_log'].diff()
wdb_sp500['GDP_log_diff2'] = wdb_sp500['GDP_log_diff'].diff()

wdp_sp500_na = wdb_sp500[wdb_sp500['GDP_log_diff2'].isna()]
wdb_sp500.dropna(inplace=True)

adf_test(wdb_sp500['GDP_log_diff2'])
```

```python
X = wdb_sp500.drop(columns=['GDP (current US$)', 'Country Name','GDP_log_diff','GDP_log', 'GDP_log_diff2'])
y = wdb_sp500['GDP_log_diff2']
```

```python
import xgboost
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled = pd.DataFrame(X_scaled, columns=X.columns)

params = {
    'objective': 'reg:squarederror',
    'eval_metric': 'rmse',
    'max_depth': 5,
    'learning_rate': 0.1
}

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

dtrain = xgboost.DMatrix(X_train, label=y_train)
dtest = xgboost.DMatrix(X_test, label=y_test)

xgb_model = xgboost.train(params, dtrain, num_boost_round=100, evals=[(dtest, 'eval')])

importance = xgb_model.get_score(importance_type='weight')
importance_df_xgb = pd.DataFrame(list(importance.items()), columns=['Feature', 'Importance'])
importance_df_xgb['Feature'] = importance_df_xgb['Feature'].apply(lambda x: str(x))
importance_df_xgb = importance_df_xgb.sort_values(by='Importance', ascending=False)


print(importance_df_xgb[importance_df_xgb['Importance']>=1])

selected_features = importance_df_xgb[importance_df_xgb['Importance'] >= 9]['Feature'].tolist()
```

```python
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt

y = wdb_sp500['GDP_log_diff2'].dropna()
X = wdb_sp500[selected_features].loc[y.index].fillna(method='ffill')

train_size = int(len(y) * 0.8)
y_train, y_test = y[:train_size], y[train_size:]
X_train, X_test = X[:train_size], X[train_size:]

model = ARIMA(y_train, exog=X_train, order=(1, 1, 1))
model_fit = model.fit()

forecast = model_fit.forecast(steps=len(y_test), exog=X_test)

rmse = mean_squared_error(y_test, forecast, squared=False)
print("ARIMAX RMSE:", rmse)

plt.figure(figsize=(10, 6))
plt.plot(y_test.index.year, y_test, label='Actual', marker = 'o')
plt.plot(y_test.index.year, forecast, label='Forecast', marker='o', color='red')
plt.xlabel('Year')
plt.ylabel('GDP_log_diff2')
plt.xticks(y_test.index.year, rotation=45)
plt.legend()
plt.title('ARIMAX Forecast')
plt.show()
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX

model = SARIMAX(y_train, exog=X_train, order=(1, 1, 1), seasonal_order=(1, 1, 1, 12))
model_fit = model.fit(disp=False)

forecast = model_fit.forecast(steps=len(y_test), exog=X_test)

rmse = mean_squared_error(y_test, forecast, squared=False)
print("SARIMAX RMSE:", rmse)

plt.figure(figsize=(10, 6))
plt.plot(y_test.index.year, y_test, label='Actual', marker = 'o')
plt.plot(y_test.index.year, forecast, label='Forecast', marker='o', color='red')
plt.legend()
plt.title('SARIMAX Forecast')
plt.show()
```

```python

from statsmodels.tsa.api import VAR
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import TimeSeriesSplit
import numpy as np
import matplotlib.pyplot as plt


data = wdb_sp500[['GDP_log_diff2'] + selected_features].dropna()

scaler = StandardScaler()
X_scaled = scaler.fit_transform(data[selected_features])
data_scaled = pd.DataFrame(X_scaled, columns=selected_features, index=data.index)
data_scaled['GDP_log_diff2'] = data['GDP_log_diff2']

tscv = TimeSeriesSplit(n_splits=5)
for train_index, test_index in tscv.split(data):
    train_data = data.iloc[train_index]
    test_data = data.iloc[test_index]


model = VAR(train_data)
model_fit = model.fit(ic='aic')

lags = model_fit.k_ar
test_input = train_data.values[-lags:]
forecast_steps = len(test_data)

forecast = model_fit.forecast(y=test_input, steps=forecast_steps)

forecast_df = pd.DataFrame(forecast, columns=data.columns, index=test_data.index)

rmse = np.sqrt(mean_squared_error(test_data['GDP_log_diff2'], forecast_df['GDP_log_diff2']))
print("RMSE on Test Data:", rmse)

plt.figure(figsize=(10, 6))
plt.plot(test_data.index.year, test_data['GDP_log_diff2'], label='Actual',marker='o')
plt.plot(test_data.index.year, forecast_df['GDP_log_diff2'], label='Forecast',marker='o', color='red')
plt.legend()
plt.title('VAR Test Forecast')
plt.show()
```

```python
future_steps = 5
future_index = pd.date_range(start=data.index[-1], periods=future_steps + 1, freq='Y')[1:]

future_forecast = model_fit.forecast(y=data.values[-lags:], steps=future_steps)

future_forecast_df = pd.DataFrame(future_forecast, columns=data.columns, index=future_index)

plt.figure(figsize=(10, 6))
plt.plot(data.index, data['GDP_log_diff2'], label='Historical Data', marker = 'o')
plt.plot(future_forecast_df.index, future_forecast_df['GDP_log_diff2'], marker = 'o', label='Future Forecast', color='green')
plt.legend()
plt.title('VAR Future Forecast')
plt.savefig('future_forecast.png')
plt.show()
```

```python

from statsmodels.tsa.api import VAR
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import TimeSeriesSplit
import numpy as np
import matplotlib.pyplot as plt

data = wdb_sp500_news_reduced[['GDP_log_diff2'] + selected_features_news].dropna()

scaler = StandardScaler()
X_scaled = scaler.fit_transform(data[selected_features_news])
data_scaled = pd.DataFrame(X_scaled, columns=selected_features_news, index=data.index)
data_scaled['GDP_log_diff2'] = data['GDP_log_diff2']

tscv = TimeSeriesSplit(n_splits=5)
for train_index, test_index in tscv.split(data):
    train_data = data.iloc[train_index]
    test_data = data.iloc[test_index]


model = VAR(train_data)
model_fit = model.fit(maxlags=5)

lags = model_fit.k_ar
test_input = train_data.values[-lags:]
forecast_steps = len(test_data)

forecast = model_fit.forecast(y=test_input, steps=forecast_steps)

forecast_df = pd.DataFrame(forecast, columns=data.columns, index=test_data.index)

rmse = np.sqrt(mean_squared_error(test_data['GDP_log_diff2'], forecast_df['GDP_log_diff2']))
print("RMSE on Test Data:", rmse)
```

```python
future_steps = 5
future_index = pd.date_range(start=data.index[-2], periods=future_steps + 1, freq='Y')[1:]

future_forecast = model_fit.forecast(y=data.values[-lags:], steps=future_steps)

future_forecast_df = pd.DataFrame(future_forecast, columns=data.columns, index=future_index)

plt.figure(figsize=(10, 6))
plt.plot(data.index, data['GDP_log_diff2'], label='Historical Data', marker = 'o')
plt.plot(future_forecast_df.index, future_forecast_df['GDP_log_diff2'], marker = 'o', label='Future Forecast', color='green')
plt.legend()
plt.title('VAR Future Forecast')
plt.show()
```
